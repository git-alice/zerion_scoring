```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import datetime
import plotly.express as px
from pathlib import Path
import time
```

```python
def get_top_address():
    return list(Path('../data/part1/').glob('*'))

def get_bottom_address():
    return list(Path('../data/part2/').glob('*'))
```

```python
full_df = pd.DataFrame()

for address in get_top_address():
    chunk = pd.read_hdf(address)
    chunk.index.name = 'transaction'
    chunk['address'] = address.stem
    chunk = chunk.reset_index().set_index(['address', 'transaction'])
    chunk['address_type'] = 'top100'
    full_df = pd.concat([chunk, full_df])
    
for address in get_bottom_address():
    chunk = pd.read_hdf(address)
    chunk.index.name = 'transaction'
    chunk['address'] = address.stem
    chunk = chunk.reset_index().set_index(['address', 'transaction'])
    chunk['address_type'] = 'bottom'
    full_df = pd.concat([chunk, full_df])
```
